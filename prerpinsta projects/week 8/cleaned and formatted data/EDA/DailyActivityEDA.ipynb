```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('dailyActivity.csv') 
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
#info about the data columns
print(data.info())
```

```python
#name of the columns in the data
print(data.columns)
```

```python
#datatypes of the columns
print(data.dtypes)
```

```python
# Summary statistics
print(data.describe())
```

```python
#count of the values in columns
print(data.count())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
#top 10 hisghest total steps covered
df_unique = data.drop_duplicates(subset='Id')
top_steps= df_unique.nlargest(10,'TotalSteps')[['Id','TotalSteps']]
top_steps
```

```python
#top 10 hisghest distance covered
top_distances = df_unique.nlargest(10, 'TotalDistance')[['Id', 'TotalDistance']]
top_distances
```

```python
# Correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Distribution of TotalSteps
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalSteps'], bins=30, kde=True)
plt.title('Distribution of TotalSteps')
plt.xlabel('TotalSteps')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between TotalSteps and Calories
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalSteps', y='Calories', data=data)
plt.title('Relationship between TotalSteps and Calories')
plt.xlabel('TotalSteps')
plt.ylabel('Calories')
plt.show()
```

```python
# Boxplot of VeryActiveMinutes, FairlyActiveMinutes, and LightlyActiveMinutes
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes']])
plt.title('Boxplot of Activity Minutes')
plt.ylabel('Minutes')
plt.show()
```

```python
# Pairplot for selected variables
sns.pairplot(data[['TotalSteps', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'Calories']])
plt.suptitle('Pairplot of Selected Variables', y=1.02)
plt.show()
```

```python
# Time Series Analysis for TotalSteps and Calories
plt.figure(figsize=(12, 6))
sns.lineplot(x='ActivityDate', y='TotalSteps', data=data, label='TotalSteps')
sns.lineplot(x='ActivityDate', y='Calories', data=data, label='Calories')
plt.title('Time Series Analysis')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xlim(0,10)
plt.legend()
plt.show()
```

```python
# Bar Plot for Activity Distances
activity_distances = data[['SedentaryActiveDistance', 'LightActiveDistance', 'VeryActiveDistance']].sum()
plt.figure(figsize=(10, 6))
activity_distances.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Activity Distances')
plt.xlabel('Activity Type')
plt.ylabel('Distance')
plt.xticks(rotation=0)
plt.show()
```

```python
# Boxplot for TotalDistance by Id
plt.figure(figsize=(10, 6))
sns.boxplot(x='Id', y='TotalDistance', data=data)
plt.title('Total Distance by Id')
plt.xlabel('Id')
plt.ylabel('Total Distance')
plt.xticks(rotation=45)
plt.show()
```

### Summary

Activity Levels: The data reveals various levels of physical activity, including very active, moderately active, lightly active, and sedentary. This categorization helps in understanding the intensity and type of activities performed by individuals.

Steps and Distance: Theres a focus on the total steps taken and the total distance covered. The highest recorded steps and distance in the dataset are 23,186 steps and 20.4 km, respectively, indicating significant physical exertion by some individuals.

Caloric Burn: The analysis suggests a correlation between the number of steps taken and calories burned. It indicates that people burn approximately 2000 to 3000 calories for every 5k to 15k steps, providing a measure of the energy expenditure associated with different levels of activity.

Preferred Activities: Light activities are the most preferred, as indicated by the data. This preference could be due to various factors, such as lifestyle choices or the ease of integrating light activities into daily routines.

The visualizations, such as the correlation matrix and distribution plots, enhance the understanding of these insights by providing a graphical representation of the data, making it easier to identify patterns and relationships. The time series analysis further adds depth by showing trends over time. Overall, the analysis on this page offers a comprehensive view of physical activity data, which can be valuable for health and fitness assessments.
