```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('MinutelyActivity.csv') 
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
#info about the data columns
print(data.info())
```

```python
#count of the values in columns
print(data.count())
```

```python
#name of the columns in the data
print(data.columns)
```

```python
#datatypes of the columns
print(data.dtypes)
```

```python
# Summary statistics
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
#top 10 hisghest Calories
df_unique = data.drop_duplicates(subset='Id')
top_cal= df_unique.nlargest(10,'Calories')[['Id','Calories']]
top_cal
```

```python
#top 10 hisghest distance covered
top_steps = df_unique.nlargest(10, 'Steps')[['Id', 'Steps']]
top_steps
```

```python
#top 10 hisghest distance covered
top_MET = df_unique.nlargest(10, 'METs')[['Id', 'METs']]
top_MET
```

```python
#top 10 hisghest distance covered
top_Intensity = df_unique.nlargest(10, 'Intensity')[['Id', 'Intensity']]
top_Intensity
```

```python
# Correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Distribution of TotalSteps
plt.figure(figsize=(10, 6))
sns.histplot(data['Steps'], bins=30, kde=True)
plt.title('Distribution of TotalSteps')
plt.xlabel('TotalSteps')
plt.ylabel('Frequency')
plt.xlim(10,200)
plt.ylim(0,500)
plt.show()
```

```python
# Relationship between TotalSteps and Calories
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Steps', y='Calories', data=data)
plt.title('Relationship between Steps and Calories')
plt.xlabel('Steps')
plt.ylabel('Calories')
plt.show()
```

```python
# Pairplot for selected variables
sns.pairplot(data[['Calories','Intensity','METs','Steps']])
plt.suptitle('Pairplot of Selected Variables', y=1.02)
plt.show()
```

```python
# Time Series Analysis for TotalSteps and Calories
plt.figure(figsize=(12, 6))
sns.lineplot(x='ActivityMinutes', y='Steps', data=data, label='Steps')
sns.lineplot(x='ActivityMinutes', y='Calories', data=data, label='Calories')
plt.title('Time Series Analysis')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xlim(0,10)
plt.legend()
plt.show()
```

```python
# Boxplot for TotalDistance by Id
plt.figure(figsize=(10, 6))
sns.boxplot(x='Id', y='METs', data=data)
plt.title('Steps by Id')
plt.xlabel('Id')
plt.ylabel('Steps')
plt.xticks(rotation=45)
plt.show()
```
